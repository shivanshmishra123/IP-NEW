# Visulisation

Data Visualisation means graphical or pictorial representation of the data using graph,chart,etc. Helps us show relationships betwween variables in a better manner. Effective communication of information to users. 

Parts of Plot- Plotting area, legend, axis label, ticks, titile etc

```python
import matplotlib.pyplot as plt
```

# Example :- Temp and Date graph

1) We have to store the data of temp and date seperately. So we use list, dictionary and all that

2)plt.plot(x,y) plots the data. Here x axis has to be date and y temp

```python
Date = ["26 Jan","27 Jan","28 Jan"]
Temp = [22,25,26]
plt.plot(Date,Temp)
plt.show()
```

# Saving as png
use plt.savefig('x.png')

# Adding extra Custmisation

```python
Date = ["26 Jan","27 Jan","28 Jan"]
Temp = [22,25,26]
plt.plot(Date,Temp)

plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Temp vs Date Graph")
plt.grid(True)
# plt.yticks(Temp)           # in the y axis , it only shows those values which are of our concern, like here 22,25,26

plt.show()
```

# Marker & Colour(k=black,m=magenta)

A marker is a symbol that represents a data values in a line chart or a scatter plot.

D for diamond, X for x filled, + for plus,* for star,P filled plus,s for square,o for circle, comma for pixel, . for point

#  Width and style

Linewidth and linestly used. Default = 1 pixel. 

Styles can be solid, dotted dashed , dashdot.

```python
height = [121.9,124.5,129.5,134.5,140]
weight=[19.7,21.3,23.5,25.9,28.5]
plt.plot(weight,height,marker = "*",color="g",linewidth=2,linestyle="dashed",markersize=15)
plt.xlabel("Weight in Kg")
plt.ylabel("Height in cm")
plt.title("Average weight with respect to average height")
plt.xticks(weight)
plt.grid(True)
plt.savefig("hi.png")
plt.show()
```
