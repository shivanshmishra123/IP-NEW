# Bar Chart

Unlike line plots, bar charts can plot strings on x axis.

If we do not specify the column name for the x parameter in the plot(), the bar plot will plot all the columns of the DataFrame with the index(row label) of DataFrame at x axis which is a numeric starting from().

```python
import pandas as pd
df= pd.read_csv("MelaSales.csv")
import matplotlib.pyplot as plt

df.plot(kind='bar',x='Day',title="Mela Sales Report")
plt.ylabel("Sales in Rs")
plt.show()
```

# Customisations in Bar Graph

We can use edgecolor,linestyle,linewidth,color of lines etc

```python
import pandas as pd
df= pd.read_csv("MelaSales.csv")
import matplotlib.pyplot as plt

df.plot(kind='barh',x='Day',title="Mela Sales Report",color=['red','yellow','purple'])
plt.ylabel("Sales in Rs")
plt.show()
```
