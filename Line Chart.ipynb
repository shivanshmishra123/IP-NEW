# Line Chart

A line plot is a graph that shows the frequency of data along a number line. It is used for continous dataset.

```python
# importing pandas as pd
import matplotlib.pyplot as plt
import pandas as pd

# dictionary of lists
dict = {"Week 1":[5000,5900,6500,3500,4000,5300,7900],"Week 2":[4000,3000,5000,5500,3000,4300,5900],"Week 3":
       [4000,4800,3500,2500,3000,5200,6000],"Day":['Monday',"Tuesday","Wednesday","Thurday","Friday","Saturday","Sunday"]}
	
df = pd.DataFrame(dict)
# saving the dataframe
df.to_csv('MelaSales.csv')

print(df)
```

# Plotting using a CSV file

```python
df = pd.read_csv("MelaSales.csv")
df.plot(kind="line",color = ['red','blue','brown'])
ticks = df.index.tolist()
plt.xticks(ticks,df.Day)
plt.show()
```
